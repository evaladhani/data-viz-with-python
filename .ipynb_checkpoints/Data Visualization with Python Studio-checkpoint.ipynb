Below we set up the dataframe with the Goodreads dataset for you. If you run this cell, you may get an error that some lines were skipped. Do not worry about that! For this studio, we want to focus on creating an effective

```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.year.value_counts()
df["year"] = pd.to_numeric(df["year"])

df.describe()
```

```python
# above_35 = titanic[titanic["Age"] > 35]
above_2010 = df[df["year"] > 2010]

above_2010.describe()
```

```python
df[["bookID"]].count()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.

#Reading the dataset
sb.set(rc={"figure.figsize":(25, 4)}) #width=25, #height=4
books_dataset = above_2010

#Creating column chart
sb.barplot(x = 'year',y = books_dataset[["bookID"]].count(), data = books_dataset, palette = "Blues")

#Adding the aesthetics
plt.title('Books by Year')
plt.xlabel('Year')
plt.ylabel('# of books published') 

#Show the plot
plt.show()

# WHY ISN'T IT SHOWING ME ANY BARS?!
```

```python
style.use('ggplot')

sb.set(rc={"figure.figsize":(25, 12)}) #width=25, #height=4

df_total=df['year'].value_counts()
df_total.plot(kind = 'bar', legend = False)
plt.title('Number of Books Published by Year',color = "black")
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Books',color = 'black')
plt.savefig('bar_vertical.png')

plt.show()
```
